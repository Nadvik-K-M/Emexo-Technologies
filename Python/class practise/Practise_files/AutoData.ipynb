```python
def null_count():
    for col in auto.columns:
        null_count  = auto[col].isnull().sum()
        if null_count != 0:
            print( col,null_count,auto[col].dtypes)
null_count()
```

```python
def fill_nulls():
    for col in auto.columns:
        null_count  = auto[col].isnull().sum()
        if null_count != 0:
            if auto[col].dtypes == 'object':
                mode_value = auto[col].mode()[0]
                auto[col] = auto[col].fillna(mode_value)
            else:
                mean_value = auto[col].mean()
                auto[col] = auto[col].fillna(mean_value)
```
